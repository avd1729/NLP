```python
import nltk
from nltk.draw.dispersion import dispersion_plot

# Example text
text = nltk.corpus.gutenberg.words('austen-emma.txt')

# Set of words to plot
words = ['Emma', 'Knightley', 'Frank', 'Jane', 'Harriet', 'Robert']

# Display the lexical dispersion plot
dispersion_plot(text, words)
```
