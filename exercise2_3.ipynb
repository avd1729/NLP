```python
import nltk
import matplotlib.pyplot as plt

from nltk.corpus import brown

# Select a category (e.g., "news")
category = "news"

# Get the words from the selected category
words = brown.words(categories=category)

# Calculate the frequency distribution of words
fdist = nltk.FreqDist(words)

# Get a list of word-frequency pairs sorted by frequency
sorted_word_freq = sorted(fdist.items(), key=lambda x: x[1], reverse=True)

# Create cumulative frequency data
cumulative_freq = [freq for word, freq in sorted_word_freq]
cumulative_freq = [sum(cumulative_freq[:i + 1]) for i in range(len(cumulative_freq))]

# Create a cumulative frequency plot
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_freq) + 1), cumulative_freq)
plt.title(f'Cumulative Frequency Plot for {category}')
plt.xlabel('Rank (Word Frequency)')
plt.ylabel('Cumulative Frequency')
plt.grid(True)
plt.show()
```
