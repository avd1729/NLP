```python
import nltk

from nltk.corpus import gutenberg

emma_text = gutenberg.raw('austen-emma.txt')

words = nltk.word_tokenize(emma_text)

output = ""
words_per_line = 10
word_count = 0

for word in words:
    if word_count < words_per_line:
        output += word + " "
        word_count += 1
    else:
        output += "\n" + word + " "
        word_count = 1 

print(output)
```

```python
sentences = nltk.sent_tokenize(emma_text)

num_sentences = len(sentences)

print(f"Number of sentences in 'Emma': {num_sentences}")
```

```python
num_tokens = len(words)

print(f"Number of tokens in 'Emma': {num_tokens}")
```

```python
from nltk import FreqDist

freq_dist = FreqDist(words)

print("Most common words and their frequencies:")
for word, frequency in freq_dist.most_common(10):
    print(f"{word}: {frequency}")
```

```python
target_word = "Emma"

word_occurrence = freq_dist[target_word]

print(f"The word '{target_word}' occurs {word_occurrence} times in 'Emma'.")
```

```python
print("Top 5 highest frequencies of words and their frequencies:")
for word, frequency in freq_dist.most_common(5):
    print(f"{word}: {frequency}")
```

```python
from nltk.text import Text

emma_text = gutenberg.words('austen-emma.txt')

emma_text_nltk = Text(emma_text)

words_to_plot = ["Emma", "Taylor", "Woodhouse"]

emma_text_nltk.dispersion_plot(words_to_plot)
```

```python
sentence = '''I Emma Woodhouse , handsome , clever , and rich 
, with a comfortable home and happy disposition , seemed 
to unite some of the best blessings of existence ; 
and had lived nearly twenty-one years in the world with 
very little to distress or vex her .'''

words = nltk.word_tokenize(sentence)

bigrams = nltk.collocations.BigramCollocationFinder.from_words(words)

common_bigrams = bigrams.ngram_fd.most_common()

for bigram, frequency in common_bigrams:
    print(f"{bigram}: {frequency} times")
```
